```python
import pandas as pd
import os

summary = []

files = [f for f in os.listdir("data/raw") if f.endswith(".csv")]

for file in files:
    df = pd.read_csv(f"data/raw/{file}")

    summary.append({
        "File": file,
        "Rows": df.shape[0],
        "Columns": df.shape[1],
        "Missing Values": df.isnull().sum().sum()
    })

summary_df = pd.DataFrame(summary)
summary_df
```

```python
fund_master = pd.read_csv("data/raw/01_fund_master.csv")
nav_history = pd.read_csv("data/raw/02_nav_history.csv")

fund_codes = set(fund_master["amfi_code"])
nav_codes = set(nav_history["amfi_code"])

missing_codes = fund_codes - nav_codes

print("Fund Master Codes:", len(fund_codes))
print("NAV History Codes:", len(nav_codes))
print("Missing Codes:", len(missing_codes))

if len(missing_codes) == 0:
    print("✅ All AMFI codes validated successfully")
else:
    print("❌ Missing Codes:")
    print(missing_codes)
```

```python
scheme_codes = [119551, 120503, 118632, 119092, 120841]

for code in scheme_codes:
    url = f"https://api.mfapi.in/mf/{code}"

    response = requests.get(url)

    if response.status_code == 200:
        data = response.json()

        df = pd.DataFrame(data["data"])

        df.to_csv(f"data/raw/nav_{code}.csv", index=False)

        print(f"Saved NAV for {code}")
    else:
        print(f"Failed for {code}")
```
