```python
# Convert date column to datetime
nav_history["date"] = pd.to_datetime(nav_history["date"])

# Sort data
nav_history = nav_history.sort_values(["scheme_code", "date"])

# Calculate daily returns
nav_history["daily_return"] = (
    nav_history.groupby("scheme_code")["nav"]
    .pct_change()
)

# Remove missing values
returns = nav_history.dropna(subset=["daily_return"])

print(returns.head())
```

```python
print(nav_history.columns.tolist())
```

```python
# Convert date column
nav_history["date"] = pd.to_datetime(nav_history["date"])

# Sort by AMFI code and date
nav_history = nav_history.sort_values(["amfi_code", "date"])

# Calculate daily return
nav_history["daily_return"] = (
    nav_history.groupby("amfi_code")["nav"]
    .pct_change()
)

returns = nav_history.dropna(subset=["daily_return"])

returns.head()
```

```python
var95 = (
    returns.groupby("amfi_code")["daily_return"]
    .quantile(0.05)
    .reset_index()
)

var95.columns = ["amfi_code", "VaR_95"]

var95.head()
```

```python
cvar = []

for code in returns["amfi_code"].unique():

    temp = returns[returns["amfi_code"] == code]

    var = temp["daily_return"].quantile(0.05)

    cvar_value = temp[temp["daily_return"] <= var]["daily_return"].mean()

    cvar.append([code, cvar_value])

cvar = pd.DataFrame(cvar, columns=["amfi_code", "CVaR"])

cvar.head()
```

```python
risk_report = pd.merge(var95, cvar, on="amfi_code")

risk_report.head()
```

```python
risk_report.to_csv("var_cvar_report.csv", index=False)

print("Saved Successfully!")
```

```python
# Convert benchmark date
benchmark["date"] = pd.to_datetime(benchmark["date"])

# Sort
benchmark = benchmark.sort_values("date")

# Benchmark daily return
benchmark["benchmark_return"] = benchmark["close_value"].pct_change()

benchmark.head()
```

```python
# Merge benchmark returns with fund returns
returns = pd.merge(
    returns,
    benchmark[["date", "benchmark_return"]],
    on="date",
    how="left"
)

returns.head()
```

```python
risk_free_rate = 0.065

returns["excess_return"] = returns["daily_return"] - (risk_free_rate / 252)

rolling_sharpe = (
    returns
    .groupby("amfi_code")["excess_return"]
    .rolling(90)
    .apply(lambda x: np.sqrt(252) * x.mean() / x.std() if x.std() != 0 else np.nan)
    .reset_index()
)

rolling_sharpe.columns = ["amfi_code", "index", "rolling_sharpe"]

rolling_sharpe.head()
```

```python
risk_free_rate = 0.065

returns["excess_return"] = returns["daily_return"] - (risk_free_rate / 252)

rolling_sharpe = (
    returns
    .groupby("amfi_code")["excess_return"]
    .rolling(90)
    .apply(lambda x: np.sqrt(252) * x.mean() / x.std() if x.std() != 0 else np.nan)
    .reset_index()
)

rolling_sharpe.columns = ["amfi_code", "index", "rolling_sharpe"]

rolling_sharpe.head()
```

```python
top5 = rolling_sharpe["amfi_code"].drop_duplicates().head(5)

plt.figure(figsize=(12,6))

for code in top5:
    temp = rolling_sharpe[rolling_sharpe["amfi_code"] == code]
    plt.plot(temp["index"], temp["rolling_sharpe"], label=str(code))

plt.title("Rolling 90-Day Sharpe Ratio")
plt.xlabel("Observation")
plt.ylabel("Sharpe Ratio")
plt.legend()

plt.savefig("rolling_sharpe_chart.png", dpi=300)

plt.show()
```

```python
# Convert transaction date
transactions["date"] = pd.to_datetime(transactions["date"])

# Extract transaction year
transactions["year"] = transactions["date"].dt.year

# Group by year
cohort = transactions.groupby("year").agg(
    Investors=("investor_id", "nunique"),
    Total_Investment=("amount", "sum"),
    Average_Investment=("amount", "mean")
).reset_index()

cohort
```

```python
print(transactions.columns.tolist())
```

```python
# Convert transaction date
transactions["transaction_date"] = pd.to_datetime(
    transactions["transaction_date"]
)

# Extract year
transactions["year"] = transactions["transaction_date"].dt.year

# Group by year
cohort = (
    transactions
    .groupby("year")
    .agg(
        Total_Investment=("amount_inr", "sum"),
        Total_Transactions=("investor_id", "count"),
        Unique_Investors=("investor_id", "nunique")
    )
    .reset_index()
)

cohort
```

```python
plt.figure(figsize=(10,5))

plt.bar(
    cohort["year"].astype(str),
    cohort["Total_Investment"] / 10000000
)

plt.title("Investor Cohort Analysis")
plt.xlabel("Year")
plt.ylabel("Investment (Crores)")

plt.savefig("investor_cohort_analysis.png", dpi=300)

plt.show()
```

```python
# SIP transactions only
sip = transactions[transactions["transaction_type"] == "SIP"].copy()

# Convert transaction date
sip["transaction_date"] = pd.to_datetime(sip["transaction_date"])

# Sort data
sip = sip.sort_values(["investor_id", "transaction_date"])

# Previous transaction date
sip["previous_date"] = sip.groupby("investor_id")["transaction_date"].shift(1)

# Gap between SIPs
sip["gap_days"] = (
    sip["transaction_date"] - sip["previous_date"]
).dt.days

sip.head()
```

```python
# Investors at risk (>35 day gap)
at_risk = sip[sip["gap_days"] > 35]

print("Total At-Risk Investors:", at_risk["investor_id"].nunique())

at_risk.head()
```

```python
# Average SIP gap
gap_report = (
    sip.groupby("investor_id")
    .agg(
        Avg_Gap_Days=("gap_days", "mean"),
        Max_Gap_Days=("gap_days", "max")
    )
    .reset_index()
)

gap_report.head()
```

```python
plt.figure(figsize=(12,5))

gap_report["Avg_Gap_Days"].hist(bins=25)

plt.title("Average SIP Gap Distribution")
plt.xlabel("Average Gap (Days)")
plt.ylabel("Number of Investors")

plt.savefig("sip_gap_distribution.png", dpi=300)

plt.show()
```

```python
# Average rolling Sharpe for each fund
fund_scores = (
    rolling_sharpe
    .groupby("amfi_code")["rolling_sharpe"]
    .mean()
    .reset_index()
)

fund_scores.columns = ["amfi_code", "avg_sharpe"]

fund_scores.head()
```

```python
def recommend_funds(risk_level):

    if risk_level.lower() == "low":
        return fund_scores.sort_values(
            "avg_sharpe",
            ascending=False
        ).head(3)

    elif risk_level.lower() == "medium":
        return fund_scores.sort_values(
            "avg_sharpe",
            ascending=False
        ).iloc[3:6]

    elif risk_level.lower() == "high":
        return fund_scores.sort_values(
            "avg_sharpe",
            ascending=False
        ).tail(3)

    else:
        return "Invalid Risk Level"
```

```python
recommend_funds("Low")
```

```python
recommendation = recommend_funds("Low")

recommendation.to_csv(
    "fund_recommendations.csv",
    index=False
)

print("Recommendation saved successfully!")
```
