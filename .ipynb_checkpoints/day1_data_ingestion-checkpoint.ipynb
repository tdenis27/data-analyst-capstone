```python
import requests

url = "https://api.mfapi.in/mf/125497"

response = requests.get(url)

print("Status Code:", response.status_code)

data = response.json()

print(data["meta"]["scheme_name"])
```

```python
import pandas as pd

nav_df = pd.DataFrame(data["data"])

nav_df.head()
```

```python
nav_df.to_csv("../data/raw/nav_125497.csv", index=False)

print("NAV data saved successfully")
```

```python
print("Shape:", nav_df.shape)

print("\nData Types:")
print(nav_df.dtypes)

print("\nFirst 5 Rows:")
print(nav_df.head())
```

```python
print("\nMissing Values:")
print(nav_df.isnull().sum())
```

```python
with open("reports/data_quality_summary.txt", "w") as f:
    f.write(f"Rows, Columns: {nav_df.shape}\n")
    f.write("\nData Types:\n")
    f.write(str(nav_df.dtypes))
    f.write("\n\nMissing Values:\n")
    f.write(str(nav_df.isnull().sum()))

print("Data Quality Summary Saved")
```
